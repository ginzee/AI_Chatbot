```python
# Start your code here!
from dotenv import load_dotenv
import os
from openai import OpenAI
load_dotenv()


# Create client
client = OpenAI()    
```

```python
# create variable "conversations" that lists all the messages in one

conversation = [{"role": "system",
            "content": "You are a pragmatic career and job-search assistant specializing in data analytics, "
            "business analytics, and early-to-mid career transitions into data roles.\n\n"
            "Your role is NOT to give generic motivational advice. Instead, you help users think "
            "clearly, make trade-offs explicit, and structure their decisions.\n\n"
            "Guiding principles:\n"
            "- Be concrete, analytical, and honest.\n"
            "- Ask clarifying questions when information is missing.\n"
            "- Explain reasoning step by step when helpful.\n"
            "- Avoid buzzwords unless the user explicitly asks for them.\n"
            "- Tailor advice to real-world constraints (job market, geography, experience level).\n\n"
            "You can help with:\n"
            "- Data analyst job search strategy\n"
            "- Resume and portfolio positioning\n"
            "- Interview preparation (technical and behavioral)\n"
            "- Career transitions into analytics\n"
            "- Evaluating job offers and role fit\n\n"
            "If a question is vague, ask one precise follow-up question before answering."
            "If a question does not relate to any of the topics listed in the 'You can help with' section, please respond with 'I'm sorry, but I can't help you with that.'"}]

user_qs = [
    "I'm trying to break into data analytics. Should I focus more on SQL and Python, or on tools like Tableau and Power BI?",
    
    "I have experience in consulting and marketing but less formal analytics experience. How should I position myself for data analyst roles?",
    
    "How do recruiters actually evaluate GitHub portfolios for data analyst positions?",
    
    "I keep seeing job descriptions asking for 3–5 years of experience. How should I approach these roles if I don't meet that exactly?",
    
    "What kinds of projects actually signal that someone is ready for a full-time data analyst role?",
    
    "How should I prepare differently for a technical interview versus a behavioral interview for analytics roles?",
    
    "I'm applying to roles in Europe and the US. How does the job market and hiring process differ for data analysts?",
    
    "At what point does continuing to upskill become procrastination instead of progress in a job search?",
    
    "How do I explain a career transition into data analytics in a way that feels credible and not defensive?",
    
    "How should I evaluate a data analyst job offer beyond just salary?"
]


for q in user_qs:
    print("user: ", q)
    user_dict = {"role": "user", "content": q}
    conversation.append(user_dict)

    # automate responses in assistant dictionary
    response = client.chat.completions.create(
        model = "gpt-4o-mini",
        messages=conversation
    )

    assistant_dict = {"role": "assistant", "content" : response.choices[0].message.content}
    conversation.append(assistant_dict)
    print("assistant: ", response.choices[0].message.content, "\n")
```

```python
# --- One-off demo: ask a random resume question ---

demo_question = "Can you rewrite this resume bullet to sound more data-analyst oriented? Bullet: 'Led marketing strategy for a startup and tracked KPIs.'"

conversation.append({"role": "user", "content": demo_question})

MODEL = "gpt-4o-mini"
TEMPERATURE = 0.0
MAX_TOKENS = 100

response = client.chat.completions.create(
    model=MODEL,
    messages=conversation,
    temperature=TEMPERATURE,
    max_completion_tokens=MAX_TOKENS
)

assistant_reply = response.choices[0].message.content
conversation.append({"role": "assistant", "content": assistant_reply})

print("\n" + "="*80)
print("DEMO QUESTION:")
print(demo_question)
print("-"*80)
print("ASSISTANT:")
print(assistant_reply)
print("="*80 + "\n")
```
